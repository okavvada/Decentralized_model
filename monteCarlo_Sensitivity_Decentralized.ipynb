```python
class Parameters():
    def __init__(self, demand, slope_index, MBR_Operational_energy_y, MBR_Operational_GHG_y,MBR_Capital_energy_y, MBR_GHG,pipe_diameter, pump_hp, pop_density_grid, miles_to_disposal, miles_transport, discount, lifetime, lifetime_treatment,street_density, pump_operating_fraction, Grinder_pump_usage, UV_rating, UV_capital_cost,UV_usage, Screen_Filter_capital_cost,Screen_Filter_energy,Grit_chamb_time, Grinder_pump_hp, storage_days, sludge_mass, percent_fertilizer,electricity_cost,excavation_energy,excavation_GHG,transport_energy,transport_GHG,steel_energy,steel_GHG,chlorine_energy,chlorine_GHG,chlorine_mass,pipe_energy_EF,pump_energy_EF, pump_GHG_EF, pipe_maint_energy, pipe_maint_GHG,Electricity_GHG_LCA):
        self.demand = demand
        self.slope_index = slope_index
        self.MBR_Operational_energy_y=MBR_Operational_energy_y
        self.MBR_Operational_GHG_y=MBR_Operational_GHG_y
        self.MBR_Capital_energy_y=MBR_Capital_energy_y
        self.MBR_GHG=MBR_GHG
        self.pipe_diameter = pipe_diameter
        self.pump_hp = pump_hp
        self.pop_density_grid = pop_density_grid
        self.miles_to_disposal = miles_to_disposal
        self.miles_transport = miles_transport
        self.discount = discount
        self.lifetime = lifetime
        self.lifetime_treatment=lifetime_treatment
        self.street_density = street_density
        self.pump_operating_fraction = pump_operating_fraction
        self.Grinder_pump_usage = Grinder_pump_usage
        self.UV_rating= UV_rating
        self.UV_capital_cost = UV_capital_cost
        self.UV_usage = UV_usage
        self.Screen_Filter_capital_cost = Screen_Filter_capital_cost
        self.Screen_Filter_energy=Screen_Filter_energy
        self.Grit_chamb_time=Grit_chamb_time
        self.Grinder_pump_hp = Grinder_pump_hp
        self.storage_days = storage_days
        self.sludge_mass = sludge_mass
        self.percent_fertilizer = percent_fertilizer
        self.electricity_cost = electricity_cost
        self.excavation_energy = excavation_energy
        self.excavation_GHG = excavation_GHG
        self.transport_energy = transport_energy
        self.transport_GHG = transport_GHG
        self.steel_energy = steel_energy
        self.steel_GHG = steel_GHG
        self.chlorine_energy = chlorine_energy
        self.chlorine_GHG = chlorine_GHG
        self.chlorine_mass = chlorine_mass
        self.pipe_energy_EF = pipe_energy_EF
        self.pump_energy_EF = pump_energy_EF
        self.pump_GHG_EF = pump_GHG_EF
        self.pipe_maint_energy = pipe_maint_energy
        self.Electricity_GHG_LCA = Electricity_GHG_LCA
        self.pipe_maint_GHG = pipe_maint_GHG
        
    def perturb(self, perturbation):
        new_demand = self.demand
        new_slope_index = self.slope_index
        new_pipe_diameter=self.pipe_diameter
        new_pump_hp=self.pump_hp
        new_MBR_Operational_energy_y=self.MBR_Operational_energy_y
        new_MBR_Operational_GHG_y=self.MBR_Operational_GHG_y
        new_MBR_Capital_energy_y=self.MBR_Capital_energy_y
        new_MBR_GHG=self.MBR_GHG
        new_pop_density_grid=self.pop_density_grid
        new_miles_to_disposal=self.miles_to_disposal
        new_miles_transport=self.miles_transport
        new_discount=self.discount
        new_lifetime = self.lifetime
        new_lifetime_treatment=self.lifetime_treatment
        new_street_density = self.street_density
        new_Grinder_pump_usage = self.Grinder_pump_usage
        new_UV_rating= self.UV_rating
        new_UV_capital_cost = self.UV_capital_cost
        new_UV_usage = self.UV_usage
        new_Grinder_pump_hp = self.Grinder_pump_hp
        new_Screen_Filter_capital_cost= self.Screen_Filter_capital_cost
        new_Screen_Filter_energy=self.Screen_Filter_energy
        new_Grit_chamb_time=self.Grit_chamb_time
        new_storage_days = self.storage_days
        new_sludge_mass = self.sludge_mass
        new_percent_fertilizer = self.percent_fertilizer
        new_pump_operating_fraction = self.pump_operating_fraction
        new_electricity_cost = self.electricity_cost
        new_excavation_energy = self.excavation_energy
        new_excavation_GHG = self.excavation_GHG
        new_transport_energy = self.transport_energy
        new_transport_GHG = self.transport_GHG
        new_steel_energy = self.steel_energy
        new_steel_GHG = self.steel_GHG
        new_chlorine_energy = self.chlorine_energy
        new_chlorine_GHG = self.chlorine_GHG
        new_chlorine_mass = self.chlorine_mass
        new_pipe_energy_EF = self.pipe_energy_EF
        new_pump_energy_EF = self.pump_energy_EF
        new_pump_GHG_EF = self.pump_GHG_EF
        new_pipe_maint_energy = self.pipe_maint_energy
        new_pipe_maint_GHG = self.pipe_maint_GHG
        new_Electricity_GHG_LCA = self.Electricity_GHG_LCA
        if (perturbation.key == 'demand'):
            new_demand+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'slope_index'):
            new_slope_index+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pipe_diameter'):
            new_pipe_diameter+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pump_hp'):
            new_pump_hp+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'MBR_Operational_energy_y'):
            new_MBR_Operational_energy_y+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'MBR_Operational_GHG_y'):
            new_MBR_Operational_GHG_y+= perturbation.delta * perturbation.direction    
        if (perturbation.key == 'MBR_Capital_energy_y'):
            new_MBR_Capital_energy_y+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'MBR_GHG'):
            new_MBR_GHG+= perturbation.delta * perturbation.direction   
        if (perturbation.key == 'pop_density_grid'):
            new_pop_density_grid+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'miles_to_disposal'):
            new_miles_to_disposal+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'miles_transport'):
            new_miles_transport+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'discount'):
            new_discount+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'lifetime'):
            new_lifetime+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'lifetime_treatment'):
            new_lifetime_treatment+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'street_density'):
            new_street_density+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'Grinder_pump_usage'):
            new_Grinder_pump_usage+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'UV_rating'):
            new_UV_rating+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'UV_capital_cost'):
            new_UV_capital_cost+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'UV_usage'):
            new_UV_usage+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'Screen_Filter_capital_cost'):
            new_Screen_Filter_capital_cost+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'Screen_Filter_energy'):
            new_Screen_Filter_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'Grit_chamb_time'):
            new_Grit_chamb_time+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'Grinder_pump_hp'):
            new_Grinder_pump_hp+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'storage_days'):
            new_storage_days+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'sludge_mass'):
            new_sludge_mass+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'percent_fertilizer'):
            new_percent_fertilizer+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pump_operating_fraction'):
            new_pump_operating_fraction+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'electricity_cost'):
            new_electricity_cost+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'excavation_energy'):
            new_excavation_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'excavation_GHG'):
            new_excavation_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'transport_energy'):
            new_transport_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'transport_GHG'):
            new_transport_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'steel_energy'):
            new_steel_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'steel_GHG'):
            new_steel_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'chlorine_energy'):
            new_chlorine_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'chlorine_GHG'):
            new_chlorine_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'chlorine_mass'):
            new_chlorine_mass += perturbation.delta * perturbation.direction
        if (perturbation.key == 'pipe_energy_EF'):
            new_pipe_energy_EF+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pump_energy_EF'):
            new_pump_energy_EF+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pump_GHG_EF'):
            new_pump_GHG_EF+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pipe_maint_energy'):
            new_pipe_maint_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pipe_maint_GHG'):
            new_pipe_maint_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'Electricity_GHG_LCA'):
            new_Electricity_GHG_LCA += perturbation.delta * perturbation.direction
                
        return Parameters(new_demand, new_slope_index, new_MBR_Operational_energy_y, new_MBR_Operational_GHG_y,new_MBR_Capital_energy_y, new_MBR_GHG, new_pipe_diameter, new_pump_hp, new_pop_density_grid, new_miles_to_disposal, new_miles_transport, new_discount, new_lifetime, new_lifetime_treatment, new_street_density, new_pump_operating_fraction, new_Grinder_pump_usage, new_UV_rating, new_UV_capital_cost,new_UV_usage, new_Screen_Filter_capital_cost, new_Screen_Filter_energy,new_Grit_chamb_time, new_Grinder_pump_hp, new_storage_days, new_sludge_mass, new_percent_fertilizer,new_electricity_cost,new_excavation_energy,new_excavation_GHG,new_transport_energy,new_transport_GHG,new_steel_energy,new_steel_GHG,new_chlorine_energy,new_chlorine_GHG,new_chlorine_mass,new_pipe_energy_EF,new_pump_energy_EF,new_pump_GHG_EF, new_pipe_maint_energy,new_pipe_maint_GHG,new_Electricity_GHG_LCA)

    
    def simulate(self):
        normDiameterList = (40, 50, 63, 75, 90, 100, 125, 160, 200, 250, 300, 350, 400, 450, 500, 600) 
        pipe_diameter=min(normDiameterList, key=lambda x:abs(x-self.pipe_diameter))
        pipe_length=People/self.pop_density_grid*self.street_density
        pipe_construction_df=pipe_construction_data[(pipe_construction_data['Material'] == (pipe_material))]
        h=pipe_construction_df.set_index('size_mm')


        pipe_weight=h.Wt_kg_m[pipe_diameter]
        pipe_excav_vol=h.Excavation_vol_m3_m[pipe_diameter]
        
        if math.sqrt(People/pop_density_grid)*1000<700:
            elevation=math.sqrt(People/self.pop_density_grid)*1000*self.slope_index
        else: 
            elevation=self.slope_index*700
        if elevation==0:
            elevation=0.1
            
        p_hp=self.pump_hp
        Pump_energy_hp=p_hp*3.6*8765/1.34*pump_operating_fraction
        hp=round(((Pump_energy_hp/3.6)*1.34/8765),2)
        if hp<=0.5:
            pump_hp=round(((Pump_energy_hp/3.6)*1.34/8765)*2,1)/2
        elif 0.5<hp<10:
            pump_hp=round(((Pump_energy_hp/3.6)*1.34/8765),0)
        elif 10<=hp<100:
            pump_hp=round(((Pump_energy_hp/3.6)*1.34/8765)/10,0)*10
        else:
            pump_hp=round(((Pump_energy_hp/3.6)*1.34/8765)*2/1000,1)*1000/2
     
        if p_hp<3:
            pump_efficiency=0.4
        elif 3<=p_hp<7:
            pump_efficiency=0.45
        elif 7<=p_hp<15:
            pump_efficiency=0.5
        elif 15<=p_hp<40:
            pump_efficiency=0.55
        elif 40<=p_hp<60:
            pump_efficiency=0.6
        else:
            pump_efficiency=0.7
    
        
        
        pipe_maint_GHG=self.pipe_maint_GHG/self.lifetime
        pipe_maint_energy=self.pipe_maint_energy/self.lifetime


        t=pump_construction_data.set_index('Rating_hp')
        pump_cost_EF=t.Cost_2012[pump_hp]

        pump_weight=t.Wt_kg[pump_hp]

        Pipe_mass=pipe_length*(pipe_weight)
        Pipe_excavation_vol=pipe_length*(pipe_excav_vol)
        
        head=elevation-0
        pressure=((pipe_velocity_2**2)*water_density/2+(head*101325/10)+(head*gravity*water_density)+(headloss*gravity*water_density))
        total_head=pressure/(water_density*gravity)
        
              
        Pipe_const_energy=pipe_length*(self.pipe_energy_EF)*(pipe_weight)/self.lifetime
        Pipe_const_GHG=Pipe_const_energy/3.6*Electricity_GHG_LCA_pipes
        
        Pipe_excavation_energy=Pipe_excavation_vol*(self.excavation_energy)/self.lifetime
        Pipe_excavation_GHG=Pipe_excavation_vol*(self.excavation_GHG)/self.lifetime
        
        Pipe_transport_energy=self.transport_energy*self.miles_transport*Pipe_mass/(1000*self.lifetime)
        Pipe_transport_GHG=self.transport_GHG*self.miles_transport*Pipe_mass/(1000*self.lifetime)
        
        Pipe_operat_energy=self.pipe_maint_energy
        Pipe_operat_GHG=self.pipe_maint_GHG
        
        Pump_operat_energy=specific_weight_water*total_head*(People*self.demand/3600)*self.pump_operating_fraction*24*365/(pump_efficiency*motor_efficiency)*3.6
        Pump_operat_GHG=Pump_operat_energy/3.6*self.Electricity_GHG_LCA
                
        pump_construction_energy=self.pump_energy_EF/self.lifetime
        pump_construction_GHG=self.pump_GHG_EF/self.lifetime
        
        pump_transport_energy=pump_weight*self.miles_transport*self.transport_energy/(1000*self.lifetime)
        pump_transport_GHG=pump_weight*self.miles_transport*self.transport_GHG/(1000*self.lifetime)
        
        tank_volume=self.storage_days*People*self.demand #m3
        tank_radius=math.sqrt(tank_volume/(math.pi*tank_height))
        tank_area=2*math.pi*tank_radius*tank_height+2*math.pi*tank_radius**2
        tank_mass_steel=round(tank_area/steel_sheet_area,0)*steel_sheet_mass
        
        tank_const_energy=tank_mass_steel*self.steel_energy/self.lifetime
        tank_const_GHG=tank_mass_steel*self.steel_GHG/self.lifetime
        
        tank_transport_energy=tank_mass_steel*self.transport_energy*self.miles_transport/(self.lifetime*1000)
        tank_transport_GHG=tank_mass_steel*self.transport_GHG*self.miles_transport/(self.lifetime*1000)
        
        MBR_Capital_energy_y=self.MBR_Capital_energy_y
        MBR_Capital_GHG_y=2.5087*People*self.demand/self.lifetime_treatment+self.MBR_GHG*People*self.demand*365
        MBR_Operational_energy_y=self.MBR_Operational_energy_y
        MBR_Operational_GHG_y=self.MBR_Operational_GHG_y
        #MBR_Operational_GHG_y=self.MBR_Operational_GHG_y
        
        chlorine_volume=chlorine_retention_time/24*People*self.demand
        chlorine_radius=math.sqrt(chlorine_volume/(math.pi*1))
        chlorine_cement_volume=((2*math.pi*chlorine_radius*1)+math.pi*chlorine_radius**2)*0.10
        
        Chlorine_manufacturing_energy_y=self.chlorine_mass*self.chlorine_energy*People*self.demand*365/1000+chlorine_cement_volume*cement_energy/self.lifetime_treatment
        Chlorine_manufacturing_GHG_y=self.chlorine_mass*self.chlorine_GHG*People*self.demand*365/1000+chlorine_cement_volume*cement_GHG/self.lifetime_treatment
        
        Chlorine_transport_energy_y=self.chlorine_mass*self.transport_energy*People*self.demand*365*self.miles_transport/1000000
        Chlorine_transport_GHG_y=self.chlorine_mass*self.transport_GHG*People*self.demand*365*self.miles_transport/1000000
        
        Chlorine_operational_energy_y=Chlorine_manufacturing_energy_y+Chlorine_transport_energy_y
        Chlorine_operational_GHG_y=Chlorine_manufacturing_GHG_y+Chlorine_transport_GHG_y
        
        Grit_chamb_volume=self.Grit_chamb_time/24*People*self.demand
        Grit_chamb_radius=math.sqrt(Grit_chamb_volume/(math.pi*1))
        Grit_chamb_cement_volume=((2*math.pi*Grit_chamb_radius*1)+math.pi*Grit_chamb_radius**2)*0.10
        
        Grinder_Capital_energy_y=self.pump_energy_EF/self.lifetime_treatment+Grit_chamb_cement_volume*cement_energy/self.lifetime_treatment
        Grinder_Capital_GHG_y=self.pump_GHG_EF/self.lifetime_treatment+Grit_chamb_cement_volume*cement_GHG/self.lifetime_treatment
        
        Grinder_Operational_energy_y=(self.Grinder_pump_hp*3.6/1.34*self.Grinder_pump_usage*365)
        Grinder_Operational_GHG_y=Grinder_Operational_energy_y/3.6*self.Electricity_GHG_LCA
        
        UV_Capital_energy_y=self.UV_rating*People*self.demand*self.UV_capital_cost*0.76*7.8/lifetime_UV
        UV_Capital_GHG_y=self.UV_rating*People*self.demand*self.UV_capital_cost*558*0.76/(lifetime_UV*1000)
        UV_Operational_energy_y=self.UV_rating*People*self.demand*self.UV_usage*365*3.6/1000
        UV_Operational_GHG_y=UV_Operational_energy_y/3.6*self.Electricity_GHG_LCA
        
        SF_Capital_energy_y=self.Screen_Filter_capital_cost*People*self.demand*7.18/self.lifetime_treatment
        SF_Capital_GHG_y=self.Screen_Filter_capital_cost*People*self.demand*493/1000
        SF_Operational_energy_y=self.Screen_Filter_energy*3.6*People*self.demand
        SF_Operational_GHG_y=SF_Operational_energy_y/3.6*self.Electricity_GHG_LCA
        
        Sludge_transport_energy_y=self.sludge_mass*People*self.demand*365*(length_centralized_m/1609)*self.transport_energy/1000+self.sludge_mass*0.2*People*self.demand*365*self.miles_to_disposal*self.transport_energy/1000
        Sludge_transport_GHG_y=self.sludge_mass*People*self.demand*365*(length_centralized_m/1609)*self.transport_GHG/1000+self.sludge_mass*0.2*People*self.demand*365*self.miles_to_disposal*self.transport_GHG/1000

        Sludge_disposal_GHG_y=landfill_GHG*self.sludge_mass*0.2*People*self.demand*365*(1-self.percent_fertilizer)/(1000)+fertilizer_GHG*self.sludge_mass*0.2*People*self.demand*365*self.percent_fertilizer/(1000)
        Sludge_thick_dewater_energy_y=((422832*(People*self.demand*264/(10**6))**0.9248)/365+(5024825*People*self.demand*264/(10**6)+39.693)/365)*3.6
        Sludge_thick_dewater_GHG_y=Sludge_thick_dewater_energy_y/3.6*self.Electricity_GHG_LCA

        
        return Output(demand, Pipe_const_energy, Pipe_excavation_energy, Pipe_transport_energy, Pipe_operat_energy, Pump_operat_energy, pump_construction_energy, pump_transport_energy, tank_const_energy, tank_transport_energy, MBR_Capital_energy_y, MBR_Operational_energy_y, Chlorine_manufacturing_energy_y, Chlorine_transport_energy_y,Chlorine_operational_energy_y, Grinder_Capital_energy_y, Grinder_Operational_energy_y,UV_Capital_energy_y, UV_Operational_energy_y, SF_Capital_energy_y, SF_Operational_energy_y, Sludge_transport_energy_y, Sludge_thick_dewater_energy_y, Pipe_const_GHG, Pipe_excavation_GHG, Pipe_transport_GHG, Pipe_operat_GHG, Pump_operat_GHG, pump_construction_GHG, pump_transport_GHG, tank_const_GHG, tank_transport_GHG, MBR_Capital_GHG_y,MBR_Operational_GHG_y, Chlorine_manufacturing_GHG_y, Chlorine_transport_GHG_y,Chlorine_operational_GHG_y,Grinder_Capital_GHG_y, Grinder_Operational_GHG_y,UV_Capital_GHG_y, UV_Operational_GHG_y, SF_Capital_GHG_y,SF_Operational_GHG_y,Sludge_transport_GHG_y,Sludge_disposal_GHG_y, Sludge_thick_dewater_GHG_y)

    
class Perturbation():
    def __init__(self, key, delta, direction):
        self.key = key
        self.delta = delta
        self.direction = direction  
        
        
class Output():
    def __init__(self, demand, Pipe_const_energy, Pipe_excavation_energy, Pipe_transport_energy, Pipe_operat_energy, Pump_operat_energy, pump_construction_energy, pump_transport_energy, tank_const_energy, tank_transport_energy, MBR_Capital_energy_y, MBR_Operational_energy_y, Chlorine_manufacturing_energy_y, Chlorine_transport_energy_y,Chlorine_operational_energy_y, Grinder_Capital_energy_y, Grinder_Operational_energy_y,UV_Capital_energy_y, UV_Operational_energy_y, SF_Capital_energy_y, SF_Operational_energy_y, Sludge_transport_energy_y, Sludge_thick_dewater_energy_y, Pipe_const_GHG, Pipe_excavation_GHG, Pipe_transport_GHG, Pipe_operat_GHG, Pump_operat_GHG, pump_construction_GHG, pump_transport_GHG, tank_const_GHG, tank_transport_GHG, MBR_Capital_GHG_y,MBR_Operational_GHG_y, Chlorine_manufacturing_GHG_y, Chlorine_transport_GHG_y,Chlorine_operational_GHG_y,Grinder_Capital_GHG_y, Grinder_Operational_GHG_y,UV_Capital_GHG_y, UV_Operational_GHG_y, SF_Capital_GHG_y,SF_Operational_GHG_y,Sludge_transport_GHG_y,Sludge_disposal_GHG_y, Sludge_thick_dewater_GHG_y):
        self.demand = demand
        self.Pipe_const_energy=Pipe_const_energy
        self.Pipe_const_GHG=Pipe_const_GHG
        self.Pipe_excavation_energy=Pipe_excavation_energy
        self.Pipe_excavation_GHG=Pipe_excavation_GHG
        self.Pipe_transport_energy=Pipe_transport_energy
        self.Pipe_transport_GHG=Pipe_transport_GHG
        self.Pipe_operat_energy=Pipe_operat_energy
        self.Pipe_operat_GHG=Pipe_operat_GHG
        self.Pump_operat_energy=Pump_operat_energy
        self.Pump_operat_GHG=Pump_operat_GHG
        self.pump_construction_energy=pump_construction_energy
        self.pump_construction_GHG=pump_construction_GHG
        self.pump_transport_energy=pump_transport_energy
        self.pump_transport_GHG=pump_transport_GHG
        self.tank_const_energy=tank_const_energy
        self.tank_const_GHG=tank_const_GHG
        self.tank_transport_energy=tank_transport_energy
        self.tank_transport_GHG=tank_transport_GHG
        self.MBR_Capital_energy_y=MBR_Capital_energy_y
        self.MBR_Capital_GHG_y=MBR_Capital_GHG_y
        self.MBR_Operational_energy_y=MBR_Operational_energy_y
        self.MBR_Operational_GHG_y=MBR_Operational_GHG_y
        self.Chlorine_manufacturing_energy_y=Chlorine_manufacturing_energy_y
        self.Chlorine_manufacturing_GHG_y=Chlorine_manufacturing_GHG_y
        self.Chlorine_transport_energy_y=Chlorine_transport_energy_y
        self.Chlorine_transport_GHG_y=Chlorine_transport_GHG_y
        self.Chlorine_operational_energy_y=Chlorine_operational_energy_y
        self.Chlorine_operational_GHG_y=Chlorine_operational_GHG_y
        self.Grinder_Capital_energy_y=Grinder_Capital_energy_y
        self.Grinder_Capital_GHG_y=Grinder_Capital_GHG_y
        self.Grinder_Operational_energy_y=Grinder_Operational_energy_y
        self.Grinder_Operational_GHG_y=Grinder_Operational_GHG_y
        self.UV_Capital_energy_y=UV_Capital_energy_y
        self.UV_Capital_GHG_y=UV_Capital_GHG_y
        self.UV_Operational_energy_y=UV_Operational_energy_y
        self.UV_Operational_GHG_y=UV_Operational_GHG_y
        self.SF_Capital_energy_y=SF_Capital_energy_y
        self.SF_Capital_GHG_y=SF_Capital_GHG_y
        self.SF_Operational_energy_y=SF_Operational_energy_y
        self.SF_Operational_GHG_y=SF_Operational_GHG_y
        self.Sludge_transport_energy_y=Sludge_transport_energy_y
        self.Sludge_transport_GHG_y=Sludge_transport_GHG_y
        self.Sludge_disposal_GHG_y=Sludge_disposal_GHG_y
        self.Sludge_thick_dewater_energy_y=Sludge_thick_dewater_energy_y
        self.Sludge_thick_dewater_GHG_y=Sludge_thick_dewater_GHG_y

        
    def total_cost(self):
        
        Pipe_capital_energy=self.Pipe_const_energy+self.Pipe_excavation_energy+self.Pipe_transport_energy
        Pipe_capital_GHG=self.Pipe_const_GHG+self.Pipe_excavation_GHG+self.Pipe_transport_GHG
        
        Pump_capital_energy=self.pump_construction_energy+self.pump_transport_energy
        Pump_capital_GHG=self.pump_construction_GHG+self.pump_transport_GHG
        
        Tank_capital_energy=self.tank_const_energy+self.tank_transport_energy
        Tank_capital_GHG=self.tank_const_GHG+self.tank_transport_GHG
        
        Total_conveyance_capital_GHG_y=Pipe_capital_GHG+Pump_capital_GHG+Tank_capital_GHG
        Total_conveyance_capital_energy_y=Pipe_capital_energy+Pump_capital_energy+Tank_capital_energy
        
        Total_conveyance_operational_GHG_y=self.Pipe_operat_GHG+self.Pump_operat_GHG
        Total_conveyance_operational_energy_y=self.Pipe_operat_energy+self.Pump_operat_energy
       
        Total_Treatment_Capital_energy=self.MBR_Capital_energy_y+self.Grinder_Capital_energy_y+self.UV_Capital_energy_y+self.SF_Capital_energy_y
        Total_Treatment_Capital_GHG=self.MBR_Capital_GHG_y+self.Grinder_Capital_GHG_y+self.UV_Capital_GHG_y+self.SF_Capital_GHG_y

        Total_Treatment_Operational_energy=self.MBR_Operational_energy_y+self.Chlorine_operational_energy_y+self.Sludge_transport_energy_y+self.Sludge_thick_dewater_energy_y+self.Grinder_Operational_energy_y+self.UV_Operational_energy_y+self.SF_Operational_energy_y
        Total_Treatment_Operational_GHG=self.MBR_Operational_GHG_y+self.Chlorine_operational_GHG_y+self.Sludge_transport_GHG_y+self.Sludge_disposal_GHG_y+self.Sludge_thick_dewater_GHG_y+self.Grinder_Operational_GHG_y+self.UV_Operational_GHG_y+self.SF_Operational_GHG_y

        Total_Energy_y=Total_conveyance_capital_energy_y+Total_conveyance_operational_energy_y+Total_Treatment_Capital_energy+Total_Treatment_Operational_energy
        Total_GHG_y=Total_conveyance_capital_GHG_y+Total_conveyance_operational_GHG_y+Total_Treatment_Capital_GHG+Total_Treatment_Operational_GHG
        Total_Energy_m3=Total_Energy_y/(People*self.demand*365)
        Total_GHG_m3=Total_GHG_y/(People*self.demand*365)
        
        return Total_Energy_m3
```
