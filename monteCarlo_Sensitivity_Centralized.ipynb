```python
class Parameters():
    def __init__(self, demand, slope_index, pipe_diameter, pump_hp, z_max_route, length_centralized_m,pop_density_grid, miles_to_disposal, miles_transport, discount, lifetime, lifetime_treatment, street_density, pump_operating_fraction, coagulation_time, flocculation_time, alum_density, alum_energy, alum_GHG, RSF_rate, RSF_depth, RSF_energy,RSF_sand_energy ,RSF_sand_GHG, cement_energy, cement_GHG, storage_days, sludge_mass, percent_fertilizer,electricity_cost,excavation_energy,excavation_GHG,transport_energy,transport_GHG,steel_energy,steel_GHG,chlorine_energy,chlorine_GHG,chlorine_mass,pipe_energy_EF,pump_energy_EF, pump_GHG_EF, pipe_maint_energy, pipe_maint_GHG,Electricity_GHG_LCA, Treatment_capital_energy_m3, Treatment_capital_GHG_m3,Treatment_operational_energy_m3,Treatment_operational_GHG_m3):
        self.demand = demand
        self.slope_index = slope_index
        self.pipe_diameter = pipe_diameter
        self.pump_hp=pump_hp
        self.z_max_route=z_max_route
        self.length_centralized_m=length_centralized_m
        self.pop_density_grid = pop_density_grid
        self.miles_to_disposal = miles_to_disposal
        self.miles_transport = miles_transport
        self.discount = discount
        self.lifetime = lifetime
        self.lifetime_treatment=lifetime_treatment
        self.street_density = street_density
        self.pump_operating_fraction = pump_operating_fraction
        self.coagulation_time = coagulation_time
        self.flocculation_time = flocculation_time
        self.alum_density = alum_density
        self.alum_energy = alum_energy
        self.alum_GHG = alum_GHG
        self.RSF_rate = RSF_rate
        self.RSF_depth=RSF_depth
        self.RSF_energy = RSF_energy
        self.RSF_sand_energy = RSF_sand_energy
        self.RSF_sand_GHG = RSF_sand_GHG
        self.cement_energy = cement_energy
        self.cement_GHG = cement_GHG
        self.storage_days = storage_days
        self.sludge_mass = sludge_mass
        self.percent_fertilizer = percent_fertilizer
        self.electricity_cost = electricity_cost
        self.excavation_energy = excavation_energy
        self.excavation_GHG = excavation_GHG
        self.transport_energy = transport_energy
        self.transport_GHG = transport_GHG
        self.steel_energy = steel_energy
        self.steel_GHG = steel_GHG
        self.chlorine_energy = chlorine_energy
        self.chlorine_GHG = chlorine_GHG
        self.chlorine_mass = chlorine_mass
        self.pipe_energy_EF = pipe_energy_EF
        self.pump_energy_EF = pump_energy_EF
        self.pump_GHG_EF = pump_GHG_EF
        self.pipe_maint_energy = pipe_maint_energy
        self.Electricity_GHG_LCA = Electricity_GHG_LCA
        self.pipe_maint_GHG = pipe_maint_GHG
        self.Treatment_capital_energy_m3 = Treatment_capital_energy_m3
        self.Treatment_capital_GHG_m3 = Treatment_capital_GHG_m3
        self.Treatment_operational_energy_m3 = Treatment_operational_energy_m3
        self.Treatment_operational_GHG_m3 = Treatment_operational_GHG_m3
        
        
        
    def perturb(self, perturbation):
        new_demand = self.demand
        new_slope_index = self.slope_index
        new_pipe_diameter=self.pipe_diameter
        new_pump_hp=self.pump_hp
        new_z_max_route=self.z_max_route
        new_length_centralized_m=self.length_centralized_m
        new_pop_density_grid=self.pop_density_grid
        new_miles_to_disposal=self.miles_to_disposal
        new_miles_transport=self.miles_transport
        new_discount=self.discount
        new_lifetime = self.lifetime
        new_lifetime_treatment=self.lifetime_treatment
        new_street_density = self.street_density
        new_coagulation_time = self.coagulation_time
        new_flocculation_time = self.flocculation_time
        new_alum_density = self.alum_density
        new_alum_energy = self.alum_energy
        new_alum_GHG = self.alum_GHG
        new_RSF_rate = self.RSF_rate
        new_RSF_depth = self.RSF_depth
        new_RSF_energy = self.RSF_energy
        new_RSF_sand_energy = self.RSF_sand_energy
        new_RSF_sand_GHG = self.RSF_sand_GHG
        new_cement_energy = self.cement_energy
        new_cement_GHG = self.cement_GHG
        new_storage_days = self.storage_days
        new_sludge_mass = self.sludge_mass
        new_percent_fertilizer = self.percent_fertilizer
        new_pump_operating_fraction = self.pump_operating_fraction
        new_electricity_cost = self.electricity_cost
        new_excavation_energy = self.excavation_energy
        new_excavation_GHG = self.excavation_GHG
        new_transport_energy = self.transport_energy
        new_transport_GHG = self.transport_GHG
        new_steel_energy = self.steel_energy
        new_steel_GHG = self.steel_GHG
        new_chlorine_energy = self.chlorine_energy
        new_chlorine_GHG = self.chlorine_GHG
        new_chlorine_mass = self.chlorine_mass
        new_pipe_energy_EF = self.pipe_energy_EF
        new_pump_energy_EF = self.pump_energy_EF
        new_pump_GHG_EF = self.pump_GHG_EF
        new_pipe_maint_energy = self.pipe_maint_energy
        new_pipe_maint_GHG = self.pipe_maint_GHG
        new_Electricity_GHG_LCA = self.Electricity_GHG_LCA
        new_Treatment_capital_energy_m3 = self.Treatment_capital_energy_m3
        new_Treatment_capital_GHG_m3 = self.Treatment_capital_GHG_m3
        new_Treatment_operational_energy_m3 = self.Treatment_operational_energy_m3
        new_Treatment_operational_GHG_m3 = self.Treatment_operational_GHG_m3
        
        if (perturbation.key == 'demand'):
            new_demand+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'slope_index'):
            new_slope_index+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pipe_diameter'):
            new_pipe_diameter+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pump_hp'):
            new_pump_hp+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'z_max_route'):
            new_z_max_route+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'length_centralized_m'):
            new_length_centralized_m+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pop_density_grid'):
            new_pop_density_grid+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'miles_to_disposal'):
            new_miles_to_disposal+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'miles_transport'):
            new_miles_transport+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'discount'):
            new_discount+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'lifetime'):
            new_lifetime+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'lifetime_treatment'):
            new_lifetime_treatment= perturbation.delta * perturbation.direction
        if (perturbation.key == 'street_density'):
            new_street_density+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'coagulation_time'):
            new_coagulation_time+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'flocculation_time'):
            new_flocculation_time+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'alum_density'):
            new_alum_density+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'alum_energy'):
            new_alum_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'alum_GHG'):
            new_alum_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'RSF_rate'):
            new_RSF_rate+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'RSF_depth'):
            new_RSF_depth+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'RSF_energy'):
            new_RSF_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'RSF_sand_energy'):
            new_RSF_sand_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'RSF_sand_GHG'):
            new_RSF_sand_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'cement_energy'):
            new_cement_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'cement_GHG'):
            new_cement_GHG+= perturbation.delta * perturbation.direction    
        if (perturbation.key == 'storage_days'):
            new_storage_days+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'sludge_mass'):
            new_sludge_mass+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'percent_fertilizer'):
            new_percent_fertilizer+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pump_operating_fraction'):
            new_pump_operating_fraction+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'electricity_cost'):
            new_electricity_cost+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'excavation_energy'):
            new_excavation_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'excavation_GHG'):
            new_excavation_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'transport_energy'):
            new_transport_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'transport_GHG'):
            new_transport_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'steel_energy'):
            new_steel_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'steel_GHG'):
            new_steel_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'chlorine_energy'):
            new_chlorine_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'chlorine_GHG'):
            new_chlorine_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'chlorine_mass'):
            new_chlorine_mass += perturbation.delta * perturbation.direction
        if (perturbation.key == 'pipe_energy_EF'):
            new_pipe_energy_EF+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pump_energy_EF'):
            new_pump_energy_EF+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pump_GHG_EF'):
            new_pump_GHG_EF+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pipe_maint_energy'):
            new_pipe_maint_energy+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'pipe_maint_GHG'):
            new_pipe_maint_GHG+= perturbation.delta * perturbation.direction
        if (perturbation.key == 'Electricity_GHG_LCA'):
            new_Electricity_GHG_LCA += perturbation.delta * perturbation.direction
        if (perturbation.key == 'Treatment_capital_energy_m3'):
            new_Treatment_capital_energy_m3 += perturbation.delta * perturbation.direction
        if (perturbation.key == 'Treatment_capital_GHG_m3'):
            new_Treatment_capital_GHG_m3 += perturbation.delta * perturbation.direction
        if (perturbation.key == 'Treatment_operational_energy_m3'):
            new_Treatment_operational_energy_m3 += perturbation.delta * perturbation.direction
        if (perturbation.key == 'Treatment_operational_GHG_m3'):
            new_Treatment_operational_GHG_m3 += perturbation.delta * perturbation.direction
                
        return Parameters(new_demand, new_slope_index, new_pipe_diameter, new_pump_hp, new_z_max_route, new_length_centralized_m,new_pop_density_grid, new_miles_to_disposal, new_miles_transport, new_discount, new_lifetime, new_lifetime_treatment, new_street_density, new_pump_operating_fraction, new_coagulation_time, new_flocculation_time, new_alum_density, new_alum_energy, new_alum_GHG, new_RSF_rate, new_RSF_depth,new_RSF_energy,new_RSF_sand_energy ,new_RSF_sand_GHG, new_cement_energy, new_cement_GHG, new_storage_days, new_sludge_mass, new_percent_fertilizer,new_electricity_cost,new_excavation_energy,new_excavation_GHG,new_transport_energy,new_transport_GHG,new_steel_energy,new_steel_GHG,new_chlorine_energy,new_chlorine_GHG,new_chlorine_mass,new_pipe_energy_EF,new_pump_energy_EF,new_pump_GHG_EF, new_pipe_maint_energy,new_pipe_maint_GHG,new_Electricity_GHG_LCA,new_Treatment_capital_energy_m3, new_Treatment_capital_GHG_m3,new_Treatment_operational_energy_m3,new_Treatment_operational_GHG_m3)

    
    def simulate(self):
        normDiameterList = (40, 50, 63, 75, 90, 100, 125, 160, 200, 250, 300, 350, 400, 450, 500, 600) 
        pipe_diameter=min(normDiameterList, key=lambda x:abs(x-self.pipe_diameter))
        pipe_construction_df=pipe_construction_data[(pipe_construction_data['Material'] == (pipe_material))]
        h=pipe_construction_df.set_index('size_mm')
        self.pipe_emissions_EF=h.Emissions_kgCO2_eq_m[pipe_diameter]
        self.pipe_energy_EF=h.Embodied_Energy_MJ_kg[pipe_diameter]
        pipe_weight=h.Wt_kg_m[pipe_diameter]
        pipe_excav_vol=h.Excavation_vol_m3_m[pipe_diameter]
        
        pipe_length=People/self.pop_density_grid*self.street_density
        pipe_length=pipe_length+self.length_centralized_m

        
        if math.sqrt(People/pop_density_grid)*1000<700:
            elevation=math.sqrt(People/self.pop_density_grid)*1000*self.slope_index
        else: 
            elevation=self.slope_index*700
        if elevation==0:
            elevation=0.1
            
        elevation=elevation+self.z_max_route

        p_hp=self.pump_hp
        Pump_energy_hp=p_hp*3.6*8765/1.34*pump_operating_fraction
        hp=round(((Pump_energy_hp/3.6)*1.34/8765),2)
        if hp<=0.5:
            pump_hp=round(((Pump_energy_hp/3.6)*1.34/8765)*2,1)/2
        elif 0.5<hp<10:
            pump_hp=round(((Pump_energy_hp/3.6)*1.34/8765),0)
        elif 10<=hp<100:
            pump_hp=round(((Pump_energy_hp/3.6)*1.34/8765)/10,0)*10
        else:
            pump_hp=round(((Pump_energy_hp/3.6)*1.34/8765)*2/1000,1)*1000/2
     
        if p_hp<3:
            pump_efficiency=0.4
        elif 3<=p_hp<7:
            pump_efficiency=0.45
        elif 7<=p_hp<15:
            pump_efficiency=0.5
        elif 15<=p_hp<40:
            pump_efficiency=0.55
        elif 40<=p_hp<60:
            pump_efficiency=0.6
        else:
            pump_efficiency=0.7
        
        pipe_maint_GHG=self.pipe_maint_GHG/self.lifetime
        pipe_maint_energy=self.pipe_maint_energy/self.lifetime
        

        t=pump_construction_data.set_index('Rating_hp')
        pump_cost_EF=t.Cost_2012[pump_hp]

        pump_weight=t.Wt_kg[pump_hp]

        Pipe_mass=pipe_length*(pipe_weight)
        Pipe_excavation_vol=pipe_length*(pipe_excav_vol)
        
        head=elevation-0
        pressure=((pipe_velocity_2**2)*water_density/2+(head*101325/10)+(head*gravity*water_density)+(headloss*gravity*water_density))
        total_head=pressure/(water_density*gravity)
        
              
        Pipe_const_energy=pipe_length*(self.pipe_energy_EF)*(pipe_weight)/self.lifetime
        Pipe_const_GHG=Pipe_const_energy/3.6*Electricity_GHG_LCA_pipes
       
        
        Pipe_excavation_energy=Pipe_excavation_vol*(self.excavation_energy)/self.lifetime
        Pipe_excavation_GHG=Pipe_excavation_vol*(self.excavation_GHG)/self.lifetime
        
        Pipe_transport_energy=self.transport_energy*self.miles_transport*Pipe_mass/(1000*self.lifetime)
        Pipe_transport_GHG=self.transport_GHG*self.miles_transport*Pipe_mass/(1000*self.lifetime)
        
        Pipe_operat_energy=self.pipe_maint_energy
        Pipe_operat_GHG=self.pipe_maint_GHG
        
        Pump_operat_energy=specific_weight_water*total_head*(People*self.demand/3600)*self.pump_operating_fraction*24*365/(pump_efficiency*motor_efficiency)*3.6
        Pump_operat_GHG=Pump_operat_energy/3.6*self.Electricity_GHG_LCA
                
        pump_construction_energy=self.pump_energy_EF/self.lifetime
        pump_construction_GHG=self.pump_GHG_EF/self.lifetime
        
        pump_transport_energy=pump_weight*self.miles_transport*self.transport_energy/(1000*self.lifetime)
        pump_transport_GHG=pump_weight*self.miles_transport*self.transport_GHG/(1000*self.lifetime)
        
        tank_volume=self.storage_days*People*self.demand #m3
        tank_radius=math.sqrt(tank_volume/(math.pi*tank_height))
        tank_area=2*math.pi*tank_radius*tank_height+2*math.pi*tank_radius**2
        tank_mass_steel=round(tank_area/steel_sheet_area,0)*steel_sheet_mass
        
        tank_const_energy=tank_mass_steel*self.steel_energy/self.lifetime
        tank_const_GHG=tank_mass_steel*self.steel_GHG/self.lifetime
        
        tank_transport_energy=tank_mass_steel*self.transport_energy*self.miles_transport/(self.lifetime*1000)
        tank_transport_GHG=tank_mass_steel*self.transport_GHG*self.miles_transport/(self.lifetime*1000)
        
        chlorine_volume=chlorine_retention_time/24*People*self.demand
        chlorine_radius=math.sqrt(chlorine_volume/(math.pi*1))
        chlorine_cement_volume=((2*math.pi*chlorine_radius*1)+math.pi*chlorine_radius**2)*0.10
                
        Chlorine_manufacturing_energy_y=self.chlorine_mass*self.chlorine_energy*People*self.demand*365/1000+chlorine_cement_volume*cement_energy/self.lifetime_treatment
        Chlorine_manufacturing_GHG_y=self.chlorine_mass*self.chlorine_GHG*People*self.demand*365/1000+chlorine_cement_volume*self.cement_GHG/self.lifetime_treatment
        
        Chlorine_transport_energy_y=self.chlorine_mass*self.transport_energy*People*self.demand*365*self.miles_transport/1000000
        Chlorine_transport_GHG_y=self.chlorine_mass*self.transport_GHG*People*self.demand*365*self.miles_transport/1000000
        
        Chlorine_operational_energy_y=Chlorine_manufacturing_energy_y+Chlorine_transport_energy_y
        Chlorine_operational_GHG_y=Chlorine_manufacturing_GHG_y+Chlorine_transport_GHG_y
        
        demand_day=People*self.demand
        
        Central_treatment_operational_energy_y=self.Treatment_operational_energy_m3*demand_day*365
        Central_treatment_capital_energy_y=self.Treatment_capital_energy_m3*demand_day*365

        Central_treatment_operational_GHG_y=self.Treatment_operational_GHG_m3*demand_day*365
        Central_treatment_capital_GHG_y=self.Treatment_capital_energy_m3*demand_day*365

        Treatment_energy_offsets_y=-(People*self.demand*365*Water_energy_intensity)
        Treatment_GHG_offsets_y=-(People*self.demand*365*Water_GHG_intensity)

        #Rapid Sand Filtration
        RSF_Area=People*self.demand/self.RSF_rate #m2
        RSF_volume=RSF_Area*self.RSF_depth #m3
        RSF_surface_area=RSF_Area+8*RSF_volume/RSF_Area

        RSF_tank_mass_steel=round(RSF_surface_area/steel_sheet_area,0)*steel_sheet_mass
        RSF_sand_mass=RSF_sand_density*RSF_volume

        # RSF tank construction
        RSF_tank_const_energy=RSF_tank_mass_steel*self.steel_energy/self.lifetime
        RSF_tank_const_GHG=RSF_tank_mass_steel*self.steel_GHG/self.lifetime

        # RSF_tank transport
        RSF_tank_transport_energy=RSF_tank_mass_steel*self.transport_energy*self.miles_transport/(self.lifetime*1000)
        RSF_tank_transport_GHG=RSF_tank_mass_steel*self.transport_GHG*self.miles_transport/(self.lifetime*1000)

        # RSF sand construction
        RSF_sand_const_energy=RSF_sand_mass*self.RSF_sand_energy/self.lifetime_treatment
        RSF_sand_const_GHG=RSF_sand_mass*self.RSF_sand_GHG/self.lifetime_treatment

        # RSF_sand transport
        RSF_sand_transport_energy=RSF_sand_mass*self.transport_energy*self.miles_transport/(self.lifetime_treatment*1000)
        RSF_sand_transport_GHG=RSF_sand_mass*self.transport_GHG*self.miles_transport/(self.lifetime_treatment*1000)

        #Coagulation & Flocculation
        Coagulation_volume=People*self.demand/24*self.coagulation_time
        Coagulation_radius=math.sqrt(Coagulation_volume/(math.pi*1))
        Flocculation_volume=People*self.demand/24*self.flocculation_time
        Flocculation_radius=math.sqrt(Flocculation_volume/(math.pi*1))
        Coagu_Flocc_cement_volume=(((Coagulation_radius+0.1)**2-(Coagulation_radius)**2)*3.14)+((Flocculation_radius+0.1)**2-(Flocculation_radius)**2)*3.14
        Coag_Flocc_Capital_energy=Coagu_Flocc_cement_volume*self.cement_energy/self.lifetime_treatment

        #Alum addition
        Alum_mass=self.alum_density*People*self.demand/1000*365
        Alum_manufacturing_energy=Alum_mass*self.alum_energy
        Alum_transport_energy=Alum_mass*self.transport_energy*self.miles_transport/1000
        Alum_manufacturing_GHG=Alum_mass*self.alum_GHG
        Alum_transport_GHG=Alum_mass*self.transport_energy*self.miles_transport/1000
        Alum_operational_energy_y=Alum_manufacturing_energy+Alum_transport_energy
        Alum_operational_GHG_y=Alum_manufacturing_GHG+Alum_transport_GHG

        #Coagulation & Flocculation Totals
        CoagFlocc_Capital_energy_y=Coagu_Flocc_cement_volume*self.cement_energy/self.lifetime_treatment
        CoagFlocc_Capital_GHG_y=Coagu_Flocc_cement_volume*self.cement_GHG/self.lifetime_treatment
        #CoagFlocc_Operational_energy_y=(13.486*(demand_day**(-0.303)))*365*demand_day+Alum_operational_energy_y
        #CoagFlocc_Operational_GHG_y=CoagFlocc_Operational_energy_y/3.6*Electricity_GHG_LCA+Alum_operational_GHG_y
        CoagFlocc_Operational_energy_y=Alum_operational_energy_y
        CoagFlocc_Operational_GHG_y=CoagFlocc_Operational_energy_y/3.6*self.Electricity_GHG_LCA+Alum_operational_GHG_y
        
        RSF_capital_energy_y=RSF_tank_const_energy+RSF_sand_const_energy+RSF_sand_transport_energy+RSF_tank_transport_energy
        RSF_capital_GHG_y=RSF_tank_const_GHG+RSF_sand_const_GHG+RSF_sand_transport_GHG+RSF_tank_transport_GHG
        RSF_operational_energy=self.RSF_energy*3.6*People*self.demand*365
        RSF_operational_GHG=RSF_operational_energy/3.6*Electricity_GHG_LCA
      
        
        return Output(demand, Pipe_const_energy, Pipe_excavation_energy, Pipe_transport_energy, Pipe_operat_energy, Pump_operat_energy, pump_construction_energy, pump_transport_energy, tank_const_energy, tank_transport_energy, Central_treatment_capital_energy_y, Central_treatment_operational_energy_y, RSF_capital_energy_y, RSF_operational_energy, CoagFlocc_Capital_energy_y, CoagFlocc_Operational_energy_y, Chlorine_manufacturing_energy_y, Chlorine_transport_energy_y,Chlorine_operational_energy_y,Pipe_const_GHG, Pipe_excavation_GHG, Pipe_transport_GHG, Pipe_operat_GHG, Pump_operat_GHG, pump_construction_GHG, pump_transport_GHG, tank_const_GHG, tank_transport_GHG, Central_treatment_capital_GHG_y, Central_treatment_operational_GHG_y, RSF_capital_GHG_y, RSF_operational_GHG, CoagFlocc_Capital_GHG_y, CoagFlocc_Operational_GHG_y, Chlorine_manufacturing_GHG_y, Chlorine_transport_GHG_y,Chlorine_operational_GHG_y)

    
    
    
class Perturbation():
    def __init__(self, key, delta, direction):
        self.key = key
        self.delta = delta
        self.direction = direction  
        
        
class Output():
    def __init__(self, demand, Pipe_const_energy, Pipe_excavation_energy, Pipe_transport_energy, Pipe_operat_energy, Pump_operat_energy, pump_construction_energy, pump_transport_energy, tank_const_energy, tank_transport_energy, Central_treatment_capital_energy_y, Central_treatment_operational_energy_y, RSF_capital_energy_y, RSF_operational_energy, CoagFlocc_Capital_energy_y, CoagFlocc_Operational_energy_y, Chlorine_manufacturing_energy_y, Chlorine_transport_energy_y,Chlorine_operational_energy_y,Pipe_const_GHG, Pipe_excavation_GHG, Pipe_transport_GHG, Pipe_operat_GHG, Pump_operat_GHG, pump_construction_GHG, pump_transport_GHG, tank_const_GHG, tank_transport_GHG, Central_treatment_capital_GHG_y, Central_treatment_operational_GHG_y, RSF_capital_GHG_y,RSF_operational_GHG, CoagFlocc_Capital_GHG_y, CoagFlocc_Operational_GHG_y, Chlorine_manufacturing_GHG_y, Chlorine_transport_GHG_y,Chlorine_operational_GHG_y):
        self.demand = demand
        self.Pipe_const_energy=Pipe_const_energy
        self.Pipe_const_GHG=Pipe_const_GHG
        self.Pipe_excavation_energy=Pipe_excavation_energy
        self.Pipe_excavation_GHG=Pipe_excavation_GHG
        self.Pipe_transport_energy=Pipe_transport_energy
        self.Pipe_transport_GHG=Pipe_transport_GHG
        self.Pipe_operat_energy=Pipe_operat_energy
        self.Pipe_operat_GHG=Pipe_operat_GHG
        self.Pump_operat_energy=Pump_operat_energy
        self.Pump_operat_GHG=Pump_operat_GHG
        self.pump_construction_energy=pump_construction_energy
        self.pump_construction_GHG=pump_construction_GHG
        self.pump_transport_energy=pump_transport_energy
        self.pump_transport_GHG=pump_transport_GHG
        self.tank_const_energy=tank_const_energy
        self.tank_const_GHG=tank_const_GHG
        self.tank_transport_energy=tank_transport_energy
        self.tank_transport_GHG=tank_transport_GHG
        self.Central_treatment_capital_energy_y=Central_treatment_capital_energy_y
        self.Central_treatment_operational_energy_y=Central_treatment_operational_energy_y
        self.RSF_capital_energy_y=RSF_capital_energy_y
        self.RSF_operational_energy=RSF_operational_energy
        self.CoagFlocc_Capital_energy_y=CoagFlocc_Capital_energy_y
        self.CoagFlocc_Operational_energy_y=CoagFlocc_Operational_energy_y
        self.Chlorine_manufacturing_energy_y=Chlorine_manufacturing_energy_y
        self.Chlorine_transport_energy_y=Chlorine_transport_energy_y
        self.Chlorine_operational_energy_y=Chlorine_operational_energy_y
        self.Central_treatment_capital_GHG_y=Central_treatment_capital_GHG_y
        self.Central_treatment_operational_GHG_y=Central_treatment_operational_GHG_y
        self.RSF_capital_GHG_y=RSF_capital_GHG_y
        self.RSF_operational_GHG=RSF_operational_GHG        
        self.CoagFlocc_Capital_GHG_y=CoagFlocc_Capital_GHG_y
        self.CoagFlocc_Operational_GHG_y=CoagFlocc_Operational_GHG_y
        self.Chlorine_manufacturing_GHG_y=Chlorine_manufacturing_GHG_y
        self.Chlorine_transport_GHG_y=Chlorine_transport_GHG_y
        self.Chlorine_operational_GHG_y=Chlorine_operational_GHG_y


        
    def total_cost(self):
        
        Pipe_capital_energy=self.Pipe_const_energy+self.Pipe_excavation_energy+self.Pipe_transport_energy
        Pipe_capital_GHG=self.Pipe_const_GHG+self.Pipe_excavation_GHG+self.Pipe_transport_GHG
        
        Pump_capital_energy=self.pump_construction_energy+self.pump_transport_energy
        Pump_capital_GHG=self.pump_construction_GHG+self.pump_transport_GHG
        
        Tank_capital_energy=self.tank_const_energy+self.tank_transport_energy
        Tank_capital_GHG=self.tank_const_GHG+self.tank_transport_GHG
        
        Total_conveyance_capital_GHG_y=Pipe_capital_GHG+Pump_capital_GHG+Tank_capital_GHG
        Total_conveyance_capital_energy_y=Pipe_capital_energy+Pump_capital_energy+Tank_capital_energy
        
        Total_conveyance_operational_GHG_y=self.Pipe_operat_GHG+self.Pump_operat_GHG
        Total_conveyance_operational_energy_y=self.Pipe_operat_energy+self.Pump_operat_energy
              
        Total_Treatment_Capital_energy=self.Central_treatment_capital_energy_y+self.RSF_capital_energy_y+self.CoagFlocc_Capital_energy_y
        Total_Treatment_Operational_energy=self.Central_treatment_operational_energy_y+self.CoagFlocc_Operational_energy_y+self.RSF_operational_energy+self.Chlorine_operational_energy_y
        Total_Treatment_Capital_GHG=self.Central_treatment_capital_GHG_y+self.RSF_capital_GHG_y+self.CoagFlocc_Capital_GHG_y
        Total_Treatment_Operational_GHG=self.Central_treatment_operational_GHG_y+self.CoagFlocc_Operational_GHG_y+self.RSF_operational_GHG+self.Chlorine_operational_GHG_y

        Total_Energy_y=Total_conveyance_capital_energy_y+Total_conveyance_operational_energy_y+Total_Treatment_Capital_energy+Total_Treatment_Operational_energy
        Total_GHG_y=Total_conveyance_capital_GHG_y+Total_conveyance_operational_GHG_y+Total_Treatment_Capital_GHG+Total_Treatment_Operational_GHG
        Total_Energy_m3=Total_Energy_y/(People*self.demand*365)
        Total_GHG_m3=Total_GHG_y/((People*self.demand*365))
        
        return Total_Energy_m3
```
